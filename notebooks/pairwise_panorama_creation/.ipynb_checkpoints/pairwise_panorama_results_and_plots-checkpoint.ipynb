```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
metrics_line1_cv2_not_blur = pd.read_csv('../../data/images_to_create_panorama/line1/results/metrics_df_line1_cv2_not_blured.csv')
metrics_line2_cv2_not_blur = pd.read_csv('../../data/images_to_create_panorama/line2/results/metrics_df_line2_cv2_not_blured.csv')
metrics_line3_cv2_not_blur = pd.read_csv('../../data/images_to_create_panorama/line3/results/metrics_df_line3_cv2_not_blured.csv')
metrics_line1_lukas_not_blur = pd.read_csv('../../data/images_to_create_panorama/line1/results/metrics_df_line1_lukas_not_blured.csv')
metrics_line2_lukas_not_blur = pd.read_csv('../../data/images_to_create_panorama/line2/results/metrics_df_line2_lukas_not_blured.csv')
metrics_line3_lukas_not_blur = pd.read_csv('../../data/images_to_create_panorama/line3/results/metrics_df_line3_lukas_not_blured.csv')
metrics_line1_complex_algorithm_not_blur = pd.read_csv('../../data/images_to_create_panorama/line1/results/metrics_df_line1_complex_algorithm_not_blured.csv')
metrics_line2_complex_algorithm_not_blur = pd.read_csv('../../data/images_to_create_panorama/line2/results/metrics_df_line2_complex_algorithm_not_blured.csv')
metrics_line3_complex_algorithm_not_blur = pd.read_csv('../../data/images_to_create_panorama/line3/results/metrics_df_line3_complex_algorithm_not_blured.csv')
```

```python
metrics_line1_cv2_not_blur.drop('Unnamed: 0', axis=1, inplace=True)
metrics_line2_cv2_not_blur.drop('Unnamed: 0', axis=1, inplace=True)
metrics_line3_cv2_not_blur.drop('Unnamed: 0', axis=1, inplace=True)
metrics_line1_lukas_not_blur.drop('Unnamed: 0', axis=1, inplace=True)
metrics_line2_lukas_not_blur.drop('Unnamed: 0', axis=1, inplace=True)
metrics_line3_lukas_not_blur.drop('Unnamed: 0', axis=1, inplace=True)
metrics_line1_complex_algorithm_not_blur.drop('Unnamed: 0', axis=1, inplace=True)
metrics_line2_complex_algorithm_not_blur.drop('Unnamed: 0', axis=1, inplace=True)
metrics_line3_complex_algorithm_not_blur.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
metrics_line1_cv2_not_blur
```

```python
metrics_cv2 = pd.concat([metrics_line1_cv2_not_blur, metrics_line2_cv2_not_blur, metrics_line3_cv2_not_blur])
metrics_lukas = pd.concat([metrics_line1_lukas_not_blur, metrics_line2_lukas_not_blur, metrics_line3_lukas_not_blur])
metrics_complex_algorithm = pd.concat([metrics_line1_complex_algorithm_not_blur, metrics_line2_complex_algorithm_not_blur, metrics_line3_complex_algorithm_not_blur])
```

# Success rate ploting

```python
sns.countplot(data=metrics_cv2, x='success').set(title='succes in stitching for cv2');
```

```python
sns.countplot(data=metrics_lukas, x='success').set(title='succes in stitching for lukas');
```

```python
sns.countplot(data=metrics_complex_algorithm, x='success').set(title='succes in stitching for complex algoritm');
```

# Counting average time

```python
avg_time_cv2 = []
avg_time_cv2_labels = []

for resolution in [240, 480, 720, 1080, 2048]:

    avg_time_cv2.append(metrics_line1_cv2_not_blur[
        (metrics_line1_cv2_not_blur['success']==1)
        & (metrics_line1_cv2_not_blur['image_resolution']==resolution)].time.mean())
    
    
    avg_time_cv2_labels.append('cv2_line1_'+str(resolution))
    
    avg_time_cv2.append(metrics_line2_cv2_not_blur[
        (metrics_line2_cv2_not_blur['success']==1)
        & (metrics_line2_cv2_not_blur['image_resolution']==resolution)].time.mean())
    
    avg_time_cv2_labels.append('cv2_line2_'+str(resolution))
    
    avg_time_cv2.append(metrics_line3_cv2_not_blur[
        (metrics_line3_cv2_not_blur['success']==1)
        & (metrics_line3_cv2_not_blur['image_resolution']==resolution)].time.mean())
    
    avg_time_cv2_labels.append('cv2_line3_'+str(resolution))
```

```python
avg_time_line1_lukas = metrics_line1_lukas_not_blur[metrics_line1_lukas_not_blur['success']==1].time.mean()
avg_time_line2_lukas = metrics_line2_lukas_not_blur[metrics_line2_lukas_not_blur['success']==1].time.mean()
avg_time_line3_lukas = metrics_line3_lukas_not_blur[metrics_line3_lukas_not_blur['success']==1].time.mean()
avg_time_lukas = [avg_time_line1_lukas, avg_time_line2_lukas, avg_time_line3_lukas]
print(avg_time_lukas)
print(min(avg_time_lukas)*60)
```

```python
avg_time_complex_algoritm = []
avg_time_complex_algoritm_labels = []

for resolution in [240, 480, 720, 1080, 2048]:

    avg_time_complex_algoritm.append(metrics_line1_complex_algorithm_not_blur[
        (metrics_line1_complex_algorithm_not_blur['success']==1)
        & (metrics_line1_complex_algorithm_not_blur['image_resolution']==resolution)].time.mean())
    
    
    avg_time_complex_algoritm_labels.append('complex_algoritm_line1_'+str(resolution))
    
    avg_time_complex_algoritm.append(metrics_line2_complex_algorithm_not_blur[
        (metrics_line2_complex_algorithm_not_blur['success']==1)
        & (metrics_line2_complex_algorithm_not_blur['image_resolution']==resolution)].time.mean())
    
    avg_time_complex_algoritm_labels.append('complex_algoritm_line2_'+str(resolution))
    
    avg_time_complex_algoritm.append(metrics_line3_complex_algorithm_not_blur[
        (metrics_line3_complex_algorithm_not_blur['success']==1)
        & (metrics_line3_complex_algorithm_not_blur['image_resolution']==resolution)].time.mean())
    
    avg_time_complex_algoritm_labels.append('complex_algoritm_line3_'+str(resolution))
```

# Time rate ploting

```python
col_map = plt.get_cmap('Paired')
plt.xticks(rotation='vertical')
plt.title('Average time for cv2')
plt.bar(avg_time_cv2_labels, avg_time_cv2, color=col_map.colors);
```

```python
col_map = plt.get_cmap('Paired')
plt.title('Average time for lukas')
plt.bar(['line1_lukas', 'line2_lukas', 'line3_lukas'], 
        avg_time_lukas, color=col_map.colors);
```

```python
col_map = plt.get_cmap('Paired')
plt.xticks(rotation='vertical')
plt.title('Average time for cv2')
plt.bar(avg_time_complex_algoritm_labels, avg_time_complex_algoritm, color=col_map.colors);
```
